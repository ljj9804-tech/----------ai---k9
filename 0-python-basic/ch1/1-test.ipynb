```python
name = "이상용"      # 문자열(str)
age = 25             # 정수(int)
height = 175.5       # 실수(float)
is_student = True    # 불리언(bool)

print(type(name), type(age), type(height), type(is_student))
# <class 'str'> <class 'int'> <class 'float'> <class 'bool'>
```

```python
x = input("숫자를 입력하세요: ")  # input은 항상 '문자열'로 받음
print(type(x))
x = int(x)   
print(type(x))                     # 정수로 형변환
print("제곱:", x * x)
```

```python
score = 85
if score >= 90:
  print("A")
elif score >= 80:
  print("B")   # 이 줄이 실행됨
else:
  print("C 이하")
```

```python
count = 3
while count > 0:
    print("카운트다운:", count)
    count -= 1
print("발사!")
```

```python
# 1부터 5까지 합
total = 0
for i in range(1, 6):   # 1,2,3,4,5
    total += i
print("합:", total)     # 합: 15
```
