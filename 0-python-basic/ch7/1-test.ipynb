```python
import pymysql

con = pymysql.connect(
    host='127.0.0.1',
    user='webuser',
    password='webuser',
    database='webdb',
    charset='utf8')
print('DB 연결 성공!')

cur = con.cursor()   # 커서 생성 (SQL 실행 통로)
```

```python
sql = """
CREATE TABLE book (
    id INT AUTO_INCREMENT PRIMARY KEY,
    title VARCHAR(200),
    pub VARCHAR(45),
    pages INT,
    author VARCHAR(100)
)
"""
cur.execute(sql)
print('테이블 생성 성공!')
```

```python
sql = "INSERT INTO book (title, pub, pages, author) VALUES (%s, %s, %s, %s)"
cur.execute(sql, ('머신러닝 입문', '길볏', 320, '박민수'))
con.commit()   # ← 없으면 저장 안 됨!
print('1개 행 추가됨')
```

```python
cur.execute('SELECT * FROM book')
rows = cur.fetchall()        # 모든 행
for row in rows:
    print(row)
# (1, '파이썬 기본서', '한빛미디어', 300, '김철수')

row = cur.fetchone()         # 한 행만
rows2 = cur.fetchmany(2)     # n개만 (페이지네이션)
```

```python
# 수정
cur.execute('UPDATE book SET pages = %s WHERE id = %s', (350, 1))
con.commit()
```

```python
# 삭제
cur.execute('DELETE FROM book WHERE id = %s', (1,))
con.commit()
```

```python
cur.close()
con.close()
```
